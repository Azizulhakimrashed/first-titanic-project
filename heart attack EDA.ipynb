```python
data=df[df["output"]==1]
data.describe()
```

```python
sns.countplot(x="output",data=df,palette="pastel")
plt.title("distribution of data set regarding to the number of people having heart attack")
plt.xlabel("output")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot(x="cp",data=df,palette="pastel")
plt.title("distribution of data set regarding to the number of people having cheast pain")
plt.xlabel("cheast pain")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot(x="output",hue="sex",data=df,palette="pastel")
plt.title("distribution of data set regarding to the sex")
plt.xlabel("output")
plt.ylabel("count")
plt.show()
```

```python
raw_data_by_age = df.groupby("age")["thalachh"].mean()
plt.figure(figsize=(11,7))
plt.plot(raw_data_by_age.index,raw_data_by_age.values)
plt.title("mean heart rate by age")
plt.xlabel("age")
plt.ylabel("mean heart rate")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.style.use("ggplot")
sns.displot(data=df,x="age",hue="sex")
plt.title("distribution of age around people who had heart attack")
plt.xlabel("age")
plt.ylabel("count")
plt.show()
```

```python
select_column=df[["age","trtbps","chol","sex"]]
```

```python
sns.pairplot(select_column,hue="sex")
```

```python
plt.figure(figsize=(10,6))
ax=sns.scatterplot(data=df,x="age",y="trtbps",hue="sex")
ax.set(xlabel="age",ylabel="resting blood pressure")
plt.title("relationship between age and resting blood pressure")
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax=sns.scatterplot(data=df,x="chol",y="thalachh",hue="sex")
ax.set(xlabel="cholestrone",ylabel="maximum heart rate")
plt.title("relationship between cholestrone and maximum heart rate")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.style.use("ggplot")
sns.displot(df["trtbps"])
plt.title("Distribution of blood pressure around people who had heart attack")
plt.xlabel("blood pressure")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.style.use("ggplot")
sns.displot(df["chol"])
plt.title("Distribution of cholestrol around people who had heart attack")
plt.xlabel("cholestrol")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.style.use("ggplot")
sns.displot(df["thalachh"])
plt.title("Distribution of heart rate around people who had heart attack")
plt.xlabel("heart rate")
plt.ylabel("count")
plt.show()
```
